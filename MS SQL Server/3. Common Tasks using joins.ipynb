## Finding missing sequence numbers

___


##### > bash p1.sh findmissingseq.sql



We have the following code:

```
use testdb

SET NOCOUNT ON

declare @id int
declare @maxid int

set @id = 1
select @maxid = max(game_rank) from console_games

create table #idseq
(
    id int
)

--Loop until max
while @id < @maxid 
begin
    insert into #idseq values(@id)

    set @id = @id + 1
end

select top 10  s.id from  #idseq s 
left join 
	console_games t 
	on s.id = t.game_rank 
 where t.game_rank is null
 order by s.id
 
drop table #idseq


```

This syntax will work provided the records start from sequence 1 onwards.

1. Set start id
2. Set the max id. This we get using aggregate max function on game_rank field, to get the last record
3. Create a temporary table that will store the id
4. Loop from the 1 to max records, and insert ids (+1)
5. We have now a table, which has ALL the ids, that is from 1 to max(last record)
6. We have our table (console_games), with is incomplete

Explaination of code below:

```
select s.id from  #idseq s 
left join 
	console_games t 
	on s.id = t.game_rank 
 where t.game_rank is null
 order by s.id

```

1. Left join will give us all the records from the persistent table(idseq table). This is our start. Why is this out start? It contains complete picture of what we after.
2. We match this with the gamerank table
3. Now comes kicker: we only want those records that does not appear in console_games table. And where are those records, they sit in the idseq table. So we want the records where the rank is null in the console_games table.




Also:

We use the ```where clause``` to exclude the records from right hand side table.

### Find duplicate Values

____

##### > bash p1.sh insertdup.sql
##### > bash p1.sh findduprows.sql


What will constitute a duplicate:

Here we do not have a primary key. So for this records we dupplicated each field in row, to classify it as a duplicate.

Intro to group by...Rememeber how ```group by```. If need to get count on say, a particular field, e.g. proceduretype, then list the field in select, add aggregate function, in our case it wil be count(#), but we also need to list the field group by clause. This will group all same procedure types, and and indicate the number of it:


```
select  proceduretype, count(*) AS duprows  
from      ProcedureHistory  
group by proceduretype  

```

Since our definition of a duplicate row, is duplicating every field, the ```group by``` should include every field to make it ```one unit```. We then use the having statement to count that newly created group.

Lets insert a duplicate:

```
insert into ProcedureHistory(PetID,ProcedureDate,ProcedureType, ProcedureSubCode)
	values("X0-8765","2016-03-18","VACCINATIONS", "05")
```


How to get a duplicate:
```
select  PetID,ProcedureDate,ProcedureType, ProcedureSubCode, count(*) AS duprows  
from         ProcedureHistory  
group by PetID,ProcedureDate,ProcedureType, ProcedureSubCode  
having      (count(*) > 1)	
```
