## Install MS SQL Server
___

URL:

https://docs.microsoft.com/en-us/sql/linux/quickstart-install-connect-docker?view=sql-server-ver15&pivots=cs1-bash


Additional:
https://docs.microsoft.com/en-us/sql/linux/sql-server-linux-configure-docker?view=sql-server-2017#buildnonrootcontainer

Pull the image:

```
sudo docker pull mcr.microsoft.com/mssql/server:2019-CU3-ubuntu-18.04
```

EDIT: Added the volume as well. But important on the host, dont create folder. Docker will created it. An MUST create as user root, otherwise wil start and exit.

Lets create the following commands

Run Image in detach
```
sudo docker run -e "ACCEPT_EULA=Y" -e "SA_PASSWORD=Cnrche001" \
    -p 1433:1433 --name sql1 \
    -u root\
    -v /home/chesney/mssql:/var/opt/mssql/data \
    -d mcr.microsoft.com/mssql/server:2019-CU3-ubuntu-18.04 
```

But I like to see if everything connect

```
sudo docker run -e "ACCEPT_EULA=Y" -e "SA_PASSWORD=Cnrche001" \
    -p 1434:1433 --name sql2 \
    -v /home/chesney/mssql:/var/opt/mssql/data \
    -d mcr.microsoft.com/mssql/server:2019-CU3-ubuntu-18.04 
```


Change Password
```
sudo docker exec -it sql1 /opt/mssql-tools/bin/sqlcmd \
   -S localhost -U SA -P "<YourStrong@Passw0rd>" \
   -Q 'ALTER LOGIN SA WITH PASSWORD="<YourNewStrong@Passw0rd>"'
```

Connect db

```
sudo docker exec -u root -it sql1 "bash"
```

Run sqlcmd 

```
/opt/mssql-tools/bin/sqlcmd -S localhost -U SA -P "<YourNewStrong@Passw0rd>"
```

Run an file
```
cat > one.sql
Use TestDB

select A.name, A.surname, 
count(A.clientid) as totalorders 
from clients A, Orders B 
where A.clientId = b.clientID 
AND A.clientid = 1 
group by A.clientid,A.name,A.surname

Go
^C
```

```
/opt/mssql-tools/bin/sqlcmd -S localhost -U SA -P "Cnrche001" -i one.sql
```
