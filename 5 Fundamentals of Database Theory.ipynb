## Relational Keys

____

Relational keys are divided into 4 categories

+ Primary Keys
+ Foreign Keys
+ Super Keys (most fundamental)
+ Candidate Keys

Super key is any COMNINATION of columns that uniquely identifies a row in a table. So using a combinaton of columns, you can get a unique row in table e.g.

```
Emp_SSN	    Emp_Number	 Emp_Name
---------   ----------   -------- 
123456789	226	 Steve
999999321	227	 Ajeet
888997212	228	 Chaitanya
777778888	229	 Robert
```
Super keys can be:

{Emp_SSN}
{Emp_Number}
{Emp_SSN, Emp_Number}
{Emp_SSN, Emp_Name}
{Emp_SSN, Emp_Number, Emp_Name}
{Emp_Number, Emp_Name}


A candidate key is superkey that cannot be reduced further and does not have redundant attributes. Aoso termd as minimal superkeys.

Onlt the following are candidate keys, as the rest are redundant:

{Emp_SSN}
{Emp_Number}

(in order to get a UNIQUE row. This is important)

Basically what key is sufficient to get a UNIQUE row.

Candidate keys has the following properties:
 + Uniqueness (by itself its unique)
 + Irreducability (cannot be simplified anymore)
 

What can be done with these candidate  keys? We can make one primary keys?

Primary key is candidate key that is used to uniquely identify tuples within a relation (table)

Where we have more than one candidate keys, only one is chosen as a primary key.


Foreign key is attribute in one relation that matches the candidate key of another relation

## Relational Database

____

Set of tables that satisfy the following set of data integrity:

+ Entity Integrity
    No duplicate rows in table (thats why we have primary/candidate keys)
    
+ Domain Integrity
    Enforces valid entries (restriction the type, format or range of values for columns)
    
+ Referential Integrity
    Cannot remove rows that is used other records.

+ User Defined Integrity
    E.g Cust should not be null. So database has to comply with our request.


Consistency and availability is emphasis.

## Functional Dependencies

____

To define dependencies between different relations (tables). This is the core for using structured data.

If you can connect several relations together, then the amount of data (for leater decision making) you get from analysing it, is greater than than looking at tables individually.

The best example of dependencies is foreign keys. This is do reduce redundancies. With proper setup of foreign keys your data will never be corrupted, and you will have less maintenance.

Identifying dependencies help up to ensure that every rows are unique identifies by primary key column

So table not only depend on each other, but also on themselves (through primary key) to maintain data integrity.

Here we, as I can surmise, look for data (columns), that are dependent on each other.

Lets say, we have the following:

Student ID, Student Name, Subject Id, Subject Result.

So, Student Name will be dependent on Student ID, and for that we can create a table

And,  Subject Result will be dependent on Subject Id,, and for that we can create a table
